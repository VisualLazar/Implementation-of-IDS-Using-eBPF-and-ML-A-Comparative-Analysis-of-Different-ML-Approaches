# Deep Learning Model (NN)

```python
import pandas as pd
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score, f1_score
from joblib import dump

# Ensure reproducibility
random_state = 42

# Clean column names
df_top.columns = df_top.columns.str.strip()

# Define features and target
X = df_top.drop(columns=['Label'])  # Features
y = df_top['Label']  # Target

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)

# Save the split data
dump((X_train, X_test, y_train, y_test), 'data_split.joblib')

# Define the hyperparameter distributions for randomized search
param_dist = {
    'hidden_layer_sizes': [(5,), (10,)],  # Very shallow model with fewer neurons
    'activation': ['relu'],
    'solver': ['adam'],
    'alpha': [0.0001],
    'learning_rate': ['constant']
}

# Initialize MLP classifier
mlp_classifier = MLPClassifier(random_state=random_state)

# Perform randomized search cross-validation
random_search = RandomizedSearchCV(estimator=mlp_classifier, param_distributions=param_dist, n_iter=10, cv=2, scoring='accuracy', random_state=random_state, n_jobs=-1)
random_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = random_search.best_params_

# Initialize MLP classifier with the best hyperparameters
best_mlp_classifier = MLPClassifier(**best_params, random_state=random_state)

# Train the classifier with the best hyperparameters
best_mlp_classifier.fit(X_train, y_train)

# Save the best MLP classifier to a file
dump(best_mlp_classifier, 'best_mlp_classifier.joblib')

# Make predictions on the test set
y_test_pred = best_mlp_classifier.predict(X_test)

# Calculate evaluation metrics for the test set
test_accuracy = accuracy_score(y_test, y_test_pred)
test_class_report = classification_report(y_test, y_test_pred, zero_division=1)
conf_matrix = confusion_matrix(y_test, y_test_pred)
precision = precision_score(y_test, y_test_pred, average='weighted', zero_division=1)
recall = recall_score(y_test, y_test_pred, average='weighted', zero_division=1)
f1 = f1_score(y_test, y_test_pred, average='weighted', zero_division=1)

# Save evaluation metrics
metrics = {
    'test_accuracy': test_accuracy,
    'test_class_report': test_class_report,
    'conf_matrix': conf_matrix,
    'precision': precision,
    'recall': recall,
    'f1': f1
}
dump(metrics, 'evaluation_metrics.joblib')

# Print evaluation metrics
print("Best Hyperparameters:", best_params)
print("Test Accuracy:", test_accuracy)
print("Test Classification Report:")
print(test_class_report)
print("Confusion Matrix:")
print(conf_matrix)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
```

### Decision Tree using PCA

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score, f1_score


# Perform PCA for dimensionality reduction
pca = PCA(n_components=0.95)  # Retain 95% of variance
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# Initialize and train the decision tree classifier
dt_classifier = DecisionTreeClassifier(random_state=42)
dt_classifier.fit(X_train_pca, y_train)

# Make predictions on the testing set
y_pred = dt_classifier.predict(X_test_pca)

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
class_report = classification_report(y_test, y_pred, zero_division=1)
conf_matrix = confusion_matrix(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted', zero_division=1)
recall = recall_score(y_test, y_pred, average='weighted', zero_division=1)
f1 = f1_score(y_test, y_pred, average='weighted', zero_division=1)

# Print evaluation metrics
print("Accuracy:", accuracy)
print("Classification Report:")
print(class_report)
print("Confusion Matrix:")
print(conf_matrix)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
```
